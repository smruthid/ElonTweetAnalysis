```python
import pandas as pd
from pathlib import Path
from textblob import TextBlob
```

```python
kaggle_df = pd.read_csv(
    Path("KaggleData/TweetsElonMusk.csv"),
    index_col='date',
    parse_dates=True,
    infer_datetime_format=True
)
kaggle_df = kaggle_df[['username', 'tweet']]
kaggle_df.head(10)
```

```python
def subjectivity_sentiment_score(tweet):
    return TextBlob(tweet).sentiment.subjectivity

def polarity_sentiment_score(tweet):
    return TextBlob(tweet).sentiment.polarity

kaggle_df['subjectivity'] = kaggle_df['tweet'].apply(subjectivity_sentiment_score)
kaggle_df['polarity'] = kaggle_df['tweet'].apply(polarity_sentiment_score)
```
