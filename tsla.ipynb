```python
import pandas as pd
from pathlib import Path
import yfinance as yf
```

```python
tsla = yf.Ticker('tsla')
tsla_df = tsla.history()
tsla_df.head()
```

```python
tsla_df = tsla.history(interval='1d', start='2010-06-01', end='2022-09-26')
tsla_df.head()
```

```python
tsla_df.drop(columns=["Open","High","Low","Volume","Dividends","Stock Splits"], inplace=True)
tsla_df.head()
```

```python
tsla_df.to_csv('tsla.csv')
```

```python
tsla_hourly_df = tsla.history(interval="1h", start="2020-10-20", end="2022-09-27")
tsla_hourly_df.to_csv('tsla_hourly.csv')
```
