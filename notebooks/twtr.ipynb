```python
#Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
import yfinance as yf
```

```python
#Pull Ticker data from yfinance
twtr = yf.Ticker('twtr')
twtr_df = twtr.history()
twtr_df.head()
```

```python
#Select history data using dates corresponding with data needed. Interval 1 Day reasearch data between June 2010 to Sept 2022
twtr_df = twtr.history(interval='1d', start='2010-06-01', end='2022-09-26')
twtr_df.head()
```

```python
#Drop columns except "Close"
twtr_df.drop(columns=["Open","High","Low","Volume","Dividends","Stock Splits"], inplace=True)
twtr_df.head()
```

```python
#Save tsla_df as csf file 
twtr_df.to_csv('twtr.csv')
```

```python
#Select history data for hourly analysis and save twtr_hourly_df as csv file
twtr_hourly_df = twtr.history(interval="1h", start="2020-10-20", end="2022-09-27")
twtr_hourly_df.to_csv('tsla_hourly.csv')
```
