```python
#Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
import yfinance as yf
```

```python
#Pull Ticker data from yfinance
amzn = yf.Ticker('amzn')
amzn_df = amzn.history()
amzn_df.head()
```

```python
#Select history data using dates corresponding with data needed. Interval 1 Day reasearch data between June 2010 to Sept 2022
amzn_df = amzn.history(interval='1d', start='2010-06-01', end='2022-09-26')
amzn_df.head()
```

```python
#Drop columns except "Close"
amzn_df.drop(columns=["Open","High","Low","Volume","Dividends","Stock Splits"], inplace=True)
amzn_df.head()
```

```python
#Save tsla_df as csf file 
amzn_df.to_csv('../csvfiles/amzn.csv')
```

```python
#Select history data for hourly analysis and save amzn_hourly_df as csv file
amzn_hourly_df = amzn.history(interval="1h", start="2020-10-20", end="2022-09-27")
amzn_hourly_df.to_csv('../csvfiles/amzn_hourly.csv')
```
