```python
import pandas as pd
from pathlib import Path
import yfinance as yf
```

```python
sp500= yf.Ticker('spy')
sp500_df= sp500.history()
sp500_df.head()
```

```python
sp500_df = sp500.history(interval='1d', start='2010-06-01', end='2022-09-26')
sp500_df.head()
```

```python
sp500_df.drop(columns=["Open","High","Low","Volume","Dividends","Stock Splits"], inplace=True)
sp500_df.head()
```

```python
sp500_df.to_csv('sp500.csv')
```

```python
sp500_hourly_df = sp500.history(interval="1h", start="2020-10-20", end="2022-09-27")
sp500_hourly_df.to_csv('sp500_hourly.csv')
```
