```python
#Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
import yfinance as yf
```

```python
#Pull Ticker data from yfinance
doge = yf.Ticker('doge-usd')
doge_df = doge.history()
doge_df.head()
```

```python
#Select history data using dates corresponding with data needed. Interval 1 Day reasearch data between June 2010 to Sept 2022
doge_df = doge.history(interval='1d', start='2010-06-01', end='2022-09-26')
doge_df.head()
```

```python
#Drop columns except "Close"
doge_df.drop(columns=["Open","High","Low","Volume","Dividends","Stock Splits"], inplace=True)
doge_df.head()
```

```python
#Save tsla_df as csf file 
doge_df.to_csv('doge.csv')
```

```python
#Select history data for hourly analysis and save doge_hourly_df as csv file
doge_hourly_df = doge.history(interval="1h", start="2020-10-20", end="2022-09-27")
doge_hourly_df.to_csv('doge_hourly.csv')
```
