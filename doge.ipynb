```python
import pandas as pd
from pathlib import Path
import yfinance as yf
```

```python
doge = yf.Ticker('doge-usd')
doge_df = doge.history()
doge_df.head()
```

```python
doge_df = doge.history(interval='1d', start='2010-06-01', end='2022-09-26')
doge_df.head()
```

```python
doge_df.drop(columns=["Open","High","Low","Volume","Dividends","Stock Splits"], inplace=True)
doge_df.head()
```

```python
doge_df.to_csv('doge.csv')
```
